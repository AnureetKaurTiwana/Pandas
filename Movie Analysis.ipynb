# 1.  Get the movies with good ratings( above 7) and the director Names?

```python
merged_df=movies.merge(directors,how='left',left_on='director_id',right_on='id')
```

```python
merged_df.head()
```

```python
merged_df.drop(['Unnamed: 0','id_y'],axis=1,inplace=True)
```

```python
merged_df.head()
```

```python
merged_df.rename(columns={'id_x':'id'},inplace=True)
```

```python
merged_df.head()
```

```python
merged_df.budget=(merged_df.budget/10000000).round(2)
```

```python
merged_df.head()
```

```python
mask=merged_df.vote_average>=7
```

```python
merged_df.loc[mask,['title','director_name','vote_average']].sort_values('vote_average',ascending=False).reset_index(drop=True)
```

# 2. Highly rated movie released after 2014?

```python
mask2=(merged_df.vote_average>=7) & (merged_df.year>2014)
```
