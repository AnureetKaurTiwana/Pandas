# Average Budget of the directors in descending order

```python
def avgfn(x):
    return x['budget'].mean()
    
merged_df.groupby('director_name').apply(avgfn).sort_values(ascending=False)
```

# Maximum Profitable Director

```python
# Most profiatble directors, five top ten directors highest profits
def profit(x):
    return x.revenue-x.budget

merged_df['profit']=merged_df.apply(profit,axis=1)
```

```python
def mx(x):
    return x.profit.max()

merged_df.groupby('director_name').apply(mx).sort_values(ascending=False)
```

# Group by Filter
  

```python
def my_func(x):
    return x.budget.max()>=10
merged_df.groupby('director_name').filter(my_func)
```

```python
def risky_func(x):
    x['risk_factor']=(x.budget>x.revenue.mean())
    return x
y=merged_df.groupby('director_name').apply(risky_func)
```

```python
y['risk_factor'].value_counts()
```
